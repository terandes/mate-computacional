```python

def factorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return factorial(n-1)*n
```

```python
def exp_maclaurin(x,n):
    suma = 0
    print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % ("i", "x", "x^i", "f(i)", "x^i/f(i)", "suma")
    for i in range(n+1):
        xalai = float(x**i)
        factoriali = float(factorial(i))
        suma += xalai/factoriali
        
        print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % (i,x,x**i,factoriali,xalai/factoriali,suma)
    return suma
```
