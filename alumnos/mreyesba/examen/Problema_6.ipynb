```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def decimalABinario(n): 
    ans = 0
    b = 1 
    assert (n <= 256 and n >= 0), "Overflow"    
    while n > 0:
        if n % 2 == 1:
            ans += b
        b *= 10
        n /= 2 
    return ans
```

```python
arr=np.zeros((256,3))
for i in range(256):
    arr[i][:]=[decimalABinario(i),i,float(i)]
```

```python
teams_list = ["Binario", "Real", "Flotante"]
row_format ="{:>15}" * (len(teams_list) + 1)
print row_format.format("", *teams_list)
for row in(arr):
    print row_format.format(" ",*row)
```

```python
arr1=np.zeros(21)
for i in range(0,21):
    arr1[i]=2**i

n, bins, patches = plt.hist(arr1, 1, normed=1, facecolor='b', alpha=0.75)

plt.xlabel(u'N',fontsize=20)
plt.ylabel('Numeros por binade para n',fontsize=20)
plt.title('Histograma de binade',fontsize=20)


plt.axis([0, 20, 0, 1000000])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid();
```
