```python
def factorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return factorial(n-1)*n
```

```python
def exp_maclaurin(x,n):
    suma =0.0
    print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % ("i", "x", "x^i", "factorial(i)", "x^i/factorial(i)", "suma")
    for i in range(n+1):
        suma += float(x**i)/factorial(i)
        print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % (i, x, x**i, factorial(i), x**i/factorial(i), suma)
    print "Resultado es " + str(suma)
    return suma
```

El problema se da principalmente con las bases negativas, pues es una serie alternante donde el error no se "incrementa" sino varia. Una posible solucion es usar leyes de exponentes y expresear: 
$$
e^{-x} =  \frac{1}{e^x}
$$

```python
def exp_maclaurin(x,n):
    suma =0.0
    print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % ("i", "x", "x^i", "factorial(i)", "x^i/factorial(i)", "suma")
    for i in range(n+1):
        suma += float(abs(x)**i)/factorial(i)
        print "%12s\t%12s\t%12s\t%12s\t%20s\t%20s" % (i, x, x**i, factorial(i), x**i/factorial(i), 1/suma)
    if x < 0:
        return 1/suma
    else:
        return suma
exp_maclaurin(-5.5,20)
```
