```python
%%file cripa_erastotenes.py
def criba(n):
    vectormatriz=np.ones((n,),dtype=bool)
    vectormatriz[0]=0
    vectormatriz[1]=0
    N_max=int (np.sqrt(len(vectormatriz)))
    for k in range (2,N_max):
        vectormatriz[2*k::k]]=False
    return np.nonzero(vectormatriz)
maximo =[10,100,200]
for i in xrange(0,len(maximo)):
    print criba_(maximo[i])
```

```python
def correcto(arr):
    assert np.size(arr) == 9
    return np.sum(arr) == 45 and np.prod(arr) == 362880
```

```python
def descomponer3x3(arr):
    assert np.shape(arr) == (9,9)
    x = True
    for i in range(3):
        for j in range(3):
            if x:
                arrAux = arr[i*3:i*3 + 3, j*3: j*3 + 3]
                x = correcto(arrAux.reshape(9,))
    return x
```

```python
def esSolucion(arr):
    assert arr.shape() == (9,9)
    for i in range(9):
        if not correcto(arr[i, :]) or not correcto(arr[:, i]):
            return False
    return descomponer3x3(arr)
```
