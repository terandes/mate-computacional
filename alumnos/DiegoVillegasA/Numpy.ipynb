### Cuadrado Magico

Usando la definición de cuadrados mágicos crea una función que reciba un arreglo e indique si es o no un cuadrado mágico.

```python
def cuadradosmagicos(arr):
    if arr.shape[0]==arr.shape[1]:
        arrc = arr
       
        if any(np.unique(arr.sum(0))) == any(np.unique(arr.sum(1))):
            return 0;
        else:
            return 1;
```
