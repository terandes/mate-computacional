```python
"""
N_max: es el número maximo en que se repite este proceso, es la raíz de la longitud del arreglo, que en este caso sería N_max = 10
bool_arr[2*j::j]=False este slice hace falso a partir del número 2*j (contando a partir del 0) y lo va haciendo cada j veces
Ejemplo j=3, entonces los elementos falsos serían el 6,9,12,15,18,etc
"""
N_max = int(np.sqrt(len(bool_arr)))
for j in range(2, N_max):
    bool_arr[2*j::j] = False
```

# Pregunta 4

El Sudoku es un pasatiempo que consiste en un arreglo 9×9 de celdas, con dígitos del 1 al 9, de tal manera que no se repiten valores en cada columna, cada renglón y en cada subarreglo de 3×3.

(a) Crea una función llamada `correcto` que reciba un `array` $1d$ de tamaño 9 e indique si está completo o no (i.e. incluye los números del 1 al 9)

```python
def correcto (m):
    valor = True
    for i in range (0,len(m)):
        arr = m[i,] 
        fake = 0
        for j in range (1,len(m)+1):
            for k in range(0,len(m)):
                if (arr[k]==j):
                    fake+=1
            if(fake!=1): 
                valor = False
            fake = 0
    return valor
```

```python
def correcto2 (m):
    valor = True
    for i in range (0,len(m)):
        arr = m[:,i:i+1] 
        fake = 0
        for j in range (1,len(m)+1):
            for k in range(0,len(m)):
                if (arr[k]==j):
                    fake+=1
            if(fake!=1): 
                valor = False
            fake = 0
    return valor
```

```python
def descomponer3x3(m):
    valor = False
    for i in xrange (0,9,3):
        for j in xrange(0,9,3):
            arr = m[i:i+3,j:j+3]
            if (correcto(arr)):
                if (correcto2(arr)):
                    valor = True  
    return valor
```

```python
def esSolucion(m):
    valor = False
    if(correcto(m)):
        if(correcto2(m)):
            if descomponer3x3(m):
                valor = True
    return valor
```
