```python
def arrToList(arr):#pasa de un arreglo a lista para hacerlo iterable
    n=len(arr)
    resp=[]
    for en in range(0,n):
        resp.append(arr[int(en)])
    return resp
```

```python
def ebitreal(a,b,c):
    re=((-1)**a)*(2**(b-3))*(1+normaliza(c))
    return re
def ebitbin(a,b,c):
    resp=str(a) + str(decimal_a_binario(b,3)) + str(decimal_a_binario(c,4))
    return resp
def normaliza(x):
    while x>1:
        x=float(x)/10
    return x
```

```python
def decimal_a_binario(entero,cifras): #Usando el algoritmo de la división, convierte un número entero a binario
    
    dividendo = entero
    residuoStr = ''
    for i in range(0, cifras):
        residuo = dividendo % 2
        dividendo = dividendo / 2
        residuoStr = str(residuo) + residuoStr
        
    return residuoStr
```

```python
def varianza(datos):
    n=len(datos)
    sumacuad=0
    for e in datos:
        sumacuad+=e**2
    suma=0
    for e in datos:
        suma+=e
    resp=(float(1)/n-1)*(sumacuad-(float(1)/n)*(suma**2))   
    return resp
```

```python
h=[10000,10001,10002]
```

```python
varianza(h)
```
